```python
from bokeh.sampledata.us_states import data as states
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColumnDataSource, ColorBar, BasicTicker, HoverTool

#del states['HI']
#del states['AK']

import pandas as pd
import numpy as np

# outlines in lat and lon?
state_xs = [states[code]["lons"] for code in states]
state_ys = [states[code]["lats"] for code in states]

# create series of COVID counts
import requests
URL = 'https://covidtracking.com/api/states'
r = requests.get(URL)
print(r)
import json
covid = pd.read_json(json.dumps(r.json()))
covid.index = covid['state']
deaths = covid['death']
deaths = deaths.fillna(0)

# merge deaths with FIPS

FIPS = pd.read_csv('states.csv')
FIPS.index = FIPS['USPS']
FIPS
FIPS['deaths'] = deaths
#FIPS['tests'] = tests

# create a series of populations
# create a column of normalized deaths

populations = pd.read_excel('state-populations.xlsx', index_col='State')
populations = populations['population']
FIPS['populations'] = populations
FIPS['normalized_deaths'] = FIPS['deaths']/FIPS['populations']
#FIPS['normalized_tests'] = FIPS['tests']/FIPS['populations']
FIPS.head()

#color_data = [FIPS[state_code] for state_code in states] 
color_data = [FIPS.loc[state_code]['normalized_deaths'] for state_code in states] 
states = [state_code for state_code in states]
population = [populations[state_code] for state_code in states]
deaths = [deaths[state_code] for state_code in states]

source = ColumnDataSource(dict(
        xs=state_xs,
        ys=state_ys,
        data=color_data,
        states=states,
        population=population,
        deaths=deaths
    )
)

hover = HoverTool(tooltips = [('State', '@states'), 
                              ('Population', '@population'),
                              ('Deaths', '@deaths')])

mapper = LinearColorMapper(palette='Viridis256', 
                           low=np.array(color_data).min(), 
                           high=np.array(color_data).max())

p = figure(title="Title", toolbar_location="left",
           plot_width=900, plot_height=500, tools=[hover])

color_bar = ColorBar(color_mapper=mapper, ticker=BasicTicker(),
                     location=(0,0))

p.add_layout(color_bar, 'right')

#p.patches(state_xs, state_ys, fill_color={'data':color_data, 'transform':mapper}, fill_alpha=1.0,
#          line_color="#884444", line_width=2, line_alpha=0.3)
p.patches('xs', 'ys', source=source, fill_color={'field':'data', 'transform':mapper}, fill_alpha=1.0,
          line_color="#884444", line_width=2, line_alpha=0.3)

output_notebook()
show(p)
```
